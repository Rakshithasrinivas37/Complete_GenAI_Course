#### Embedding using Ollama

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="llama3.2:1b", base_url="http://host.docker.internal:11434"
)
)
embeddings
```

```python
embeds_text = embeddings.embed_documents([
    "The sun is bright.",
    "She reads a book."
])

embeds_text
```

```python
embeddings.embed_query("What is she doing?")
```

```python
## Use other embedding models

embeddings = OllamaEmbeddings(model="mxbai-embed-large", base_url="http://host.docker.internal:11434")
text="Hi, How are you?"
query_result = embeddings.embed_query(text)

query_result
```
