#### Message prompts

Message types:
1. System message: Tells model how to behave
2. Human message: user input and interactions with the model
3. AI message: Responses generated by the model
4. Tool message: output of tool calls

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are helpful in providing the good response"),
    HumanMessage("Provide response for what is AI")
]

response = model.invoke(messages)
print(response.content)
```

```python
system_prompt = """
    You are an expert Python developer with deep knowledge of Python 3.x, including standard libraries, popular frameworks (like Flask, Django, FastAPI, pandas, NumPy), and best practices for clean, efficient, and maintainable code. 

    Your tasks include:
    - Writing Python code snippets, scripts, and functions that are correct and optimized.
    - Debugging code and explaining errors clearly.
    - Suggesting Pythonic solutions and following PEP 8 coding standards.
    - Explaining your code in plain English when needed.
    - Providing examples and guidance for libraries, data structures, algorithms, and best practices.

    Always assume the user wants working Python code unless they specify otherwise. Keep explanations concise but informative, and prioritize clarity, correctness, and efficiency.
    """

messages = [
    system_prompt,
    HumanMessage("Give me a code to print fibonacci series")
]

response = model.invoke(messages)
print(response.content)
```

```python
##Create AI message
ai_message = AIMessage("I'd be happy to help you with that question!")

## Add to conversation history
messages = [
    SystemMessage("You are an helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_message,
    HumanMessage("Generate a simple hello world python code")
]

response = model.invoke(messages)
response.content
```

```python
from langchain.messages import ToolMessage

# After a model makes a tool call
# (Here, we demonstrate manually creating the messages for brevity)
ai_message = AIMessage(
    content=[],
    tool_calls=[{
        "name": "get_weather",
        "args": {"location": "San Francisco"},
        "id": "call_123"
    }]
)

# Execute tool and create result message
weather_result = "Sunny, 72¬∞F"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"  # Must match the call ID
)

# Continue conversation
messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,  # Model's tool call
    tool_message,  # Tool execution result
]
response = model.invoke(messages)  # Model processes the result
response
```
