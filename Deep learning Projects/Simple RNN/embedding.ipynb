```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = ['Transformers process text using attention.',
'Words are converted into numerical vectors.',
'Models learn meaning from large data.',
'Embeddings represent words in vector form.',
'Attention helps models understand context.',
'Neural networks improve predictions over time.'
]
```

```python
## Define Vocab size
vocab_size = 10000
```

```python
one_hot_representation = [one_hot(words, vocab_size) for words in sentences]
one_hot_representation
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential

import numpy as np
```

```python
## Pad the sentences to make it to equal length
sent_length = 10
embedded_docs = pad_sequences(one_hot_representation, padding='post', maxlen=sent_length)
embedded_docs
```

```python
embedded_docs[0]
```

```python
## Feature representation

dim=15
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=dim, input_shape=(sent_length, )))
model.compile('adam', 'mse')

model.summary()
```

```python
model.predict(np.array([embedded_docs[0]]))
```
