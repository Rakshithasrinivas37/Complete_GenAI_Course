```python
##Lambda function

addition = lambda a,b:a+b

##To call 
print(addition(2, 3))
```

```python
##Map function 

def square(num):
    return num * num

numbers = [1,2,3,4,5]
list(map(square, numbers))
```

```python
##Lambda function with map

numbers = [1,2,3,4,5]

list(map(lambda num:num * num, numbers))
```

```python
##Map multiple iterables

num1 = [1,2,3]
num2 = [4,5,6]

list(map(lambda a,b:a + b, num1, num2))
```

```python
##To convert string to integer

text = ['1', '2']
list(map(int, text))

##To convert lower case into upper case

text = ["app", "ball"]
list(map(str.upper, text))
```

```python
##Filter function

def even(num):
    if num%2==0:
        return True
    
lst = [1,2,3,4,5,6,7,8,9]
list(filter(even, lst))
```

```python
##Regular expression

import re

pattern=r'\d+'

text = "Hi, I am XYZ, 5436 is my number"
match = re.search(pattern, text)
print(match.group())

text = "IP: 192.168.1.10"
new_text = re.sub(r"\d+\.\d+\.\d+\.\d+", "X.X.X.X", text)
print(new_text)
```
