#### Stemming: Reduce words to their root(stem) from. IT removes suffix from words to get a common root.

```python
import nltk
nltk.download('punkt_tab')

from nltk.stem import PorterStemmer

words = ['Playful', 'Playing', 'Playly', 'eating', 'eatable']

stemming = PorterStemmer()

for word in words:
    print(f"{word}----->{stemming.stem(word)}")

```

```python
#Regexpstemmer

from nltk.stem import RegexpStemmer

stemming = RegexpStemmer('ing$|s$|e$|able$', min=4)

for word in words:
    print(f"{word}----->{stemming.stem(word)}")
```

```python
#Snowball stemmer

from nltk.stem import SnowballStemmer

stemming = SnowballStemmer(language="english")

for word in words:
    print(f"{word}----->{stemming.stem(word)}")
```
