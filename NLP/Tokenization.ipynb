#### Tokenization: Process of converting text into smaller units called tokens

```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize

corpus = """Tokenization is the process of breaking text into smaller parts called tokens.
These tokens help computers understand human language. Each token can be a word, part of a word, or a symbol."""

# Converting paragraph into sentences or dictionary
sentences=sent_tokenize(corpus)
print(sentences)

```

```python
#Converting sentences into words

from nltk.tokenize import word_tokenize

for sentence in sentences:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import word_tokenize

word_tokenize(corpus)
```

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

tokenizer.tokenize(corpus)
```
